```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print("numpy_array:", arr)

arr.shape
```

```python
arr2 = np.array([[1,2,3,4], [5, 6, 7, 8]])
print("numpy_array:", arr)

arr2.shape
```

```python
# dtype()
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape()
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print("shaped_array:", b)
```

```python
# astype()
a = np.array([1, 2, 3])
b = a.astype(float)
print("float:", b)
```

```python
# np.zeros(), np.ones()
z = np.zeros((2, 3))
o = np.ones((2, 3))

print("zeros:", z)
print("ones:", o)
```

```python
# ndim
a = np.array([[1, 2], [3, 4]])
print("ndim:", a.ndim)
```

```python
# arange()
a = np.arange(1, 6)
print("aramge:", a)
```

```python
# sum(), mean(), max(), min()

a = np.array([1, 2, 3, 4])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten()
a = np.array([[1, 2], [3, 4]])
print('flatten:', a.flatten())
```

```python
# T (Transpose)
a = np.array([[1, 2], [3, 4]])
print("transpose:", a.T)
```

```python
# dot() @
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print("dot/multiplication:", a.dot(b))
```

```python
# random.rand()
a = np.random.rand(2, 3) # => 0.0 and 1.0 | 0.0 -> included / 1.0 -> excluded
print("random rand:", a)
```
